```python
import numpy as np
from time import time
```

```python
# Variables.
#
n_hidden = 10
n_in = 10
n_out = 10
n_sample = 3000
```

```python
# Hyperparameters.
#
learning_rate = 0.01
momentum = 0.9
```

```python
RND_SEED = 4118956
np.random.seed(RND_SEED)
```

```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
```

```python
def tanh_prime(x):
    return 1 - np.tanh(x) ** 2
```

```python
def train(x, t, V, W, bv, bw):
    '''
    x - input data
    t - transpose
    V - hidden layer one 
    W - hidden layer two
    bv, bw - biases (one per each layer)
    '''
    
    # Forward propagation.
    #
    A = np.dot(x, V) + bv
#     print('A: \n{}\n\n'.format(A))
    Z = np.tanh(A)
#     print('Z: \n{}\n\n'.format(Z))
    
    B = np.dot(Z, V) + bw
#     print('B: \n{}\n\n'.format(B))
    Y = sigmoid(B)
#     print('Y: \n{}\n\n'.format(Y))
    
    # Back propagation.
    #
    Ew = Y - t
#     print('Ew: \n{}\n\n'.format(Ew))
    Ev = tanh_prime(A) * np.dot(W, Ew)
#     print('Ev: \n{}\n\n'.format(Ev))
    
    # Predict loss.
    #
    dW = np.outer(Z, Ew)
#     print('dW: \n{}\n\n'.format(dW))
    dV = np.outer(x, Ev)
#     print('dV: \n{}\n\n'.format(dV))
    
    # Actual loss, cross-entropy (because here we have 
    # classification problem).
    #
    loss = -np.mean(t * np.log(Y) + (1 - t) * np.log(1 - Y))
    print(np.log(Y))
    # PROBLEM: np.log(Y) is zeros!
    
    return loss, (dV, dW, Ev, Ew)
```

```python
def predict(x, V, W, bv, bw):
    A = np.dot(x, V) + bv
    B = np.dot(np.tanh(A), W) + bw
    return (sigmoid(B) > .5).astype(int)
```

```python
# Create layers.
#

V = np.random.normal(scale=.1, size=(n_in, n_hidden))
W = np.random.normal(scale=.1, size=(n_hidden, n_out))
```

```python
bv = np.zeros(n_hidden)
bw = np.zeros(n_out)
```

```python
params = [V, W, bv, bw]
```

```python
# Generate data.
#
X = np.random.binomial(1, .5, (n_sample, n_in))
T = X ^ 1
```

```python
# Training time!
#
for epoch in range(15):
    err = []
    upd = [0] * len(params)
    
    t0 = time()
    
    # For each data point, update our weights.
    #
    for i in range(X.shape[0]):
        loss, grad = train(X[i], T[i], *params)
        
        # Update loss.
        #
        for j in range(len(params)):
            params[j] -= upd[j]
        
        for j in range(len(params)):
            upd = learning_rate * grad[j] + momentum * upd[j]
            
        err.append(loss)
    print('Epoch: {epoch}, Loss: {loss}, time: {time:.4}'.format(epoch=epoch, loss=np.mean(err), time=time() - t0))

x = np.random.binomial(1, .5, n_in)
print('XOR prediction')
print(x)
print(predict(x, *params))
```
